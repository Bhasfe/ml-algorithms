# Support Vector Machines for Classification
In this notebook, we are going to work on the breast cancer dataset which is available in sklearn. The data set includes observations labeled as malignant (harmful) or benign (not harmful) cancer type. We'll try to classify labels using support vector machine.

### What is SVM ?
Support Vector Machines are mainly used for classification problems in Machine Learning. They can also be used for regression problems (you can find in  this repository). A support vector machine tries to find best "hyperplane" which separates different classes. Following figure shows that a hyperplane whose margin is maximized.

<img src="svm.png" width="300px" height="300px" align="left" />

If the classes are not linearly separable. The kernel trick can be applied to make hyperplane. Kernel trick stands for applying the different kernel functions for multidimensional space to get desired dimensions (for example 2D to 3D ). The most populer kernels are RBF (radial basis function), polynomial and linear. Following figure shows us how kernel trick can be applied for 2D space to 3D space transformation

In this notebook we'll use 'linear' as the kernel

## 1. First Steps

Firstly, we need to import necessary libraries/packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# We'll use a built-in dataset
from sklearn import datasets

# import Support Vector Classifier
from sklearn.svm import SVC

# for splitting the data into train and test sets
from sklearn.model_selection import train_test_split

# for testing the model
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

# I will keep the resulting plots
%matplotlib inline

# Enable Jupyter Notebook's intellisense
%config IPCompleter.greedy=True
```

Secondly, load the breast cancer dataset from sklearn

```python
# Load the data set
breast_cancer = datasets.load_breast_cancer()

# make features and target DataFrame s to explore them easily
X = pd.DataFrame(breast_cancer.data,columns=breast_cancer.feature_names)
y = pd.DataFrame(breast_cancer.target,columns=["Label"])
```

## 2. Explore the Data

Now we can try to understand our dataset

```python
# Display first 5 rows of DataFrames
display(X.head())
display(y.head())

# Print label names
print("Labels:",breast_cancer.target_names)

# Visualize some 2-D features to see patterns 
sns.scatterplot(x='mean radius',y='mean texture',data=X,hue=y['Label']),
plt.title("Mean texture vs Mean Radius")
plt.show()
sns.scatterplot(x='mean perimeter',y='mean area',data=X,hue=y['Label'])
plt.title("Mean area vs Mean perimeter ")
plt.show()
```

Let's apply dimensionality reduction with t-SNE (t-distributed stochastic neighbor embedding) to see whole picture in 2-D space

```python
# import TSNE
from sklearn.manifold import TSNE

# fit and transform the TSNE model
tsne = TSNE(learning_rate =  50)
tsne_f = tsne.fit_transform(breast_cancer.data)

# Create a new DataFrame to store reduced features
df = pd.DataFrame({'x':tsne_f[:,0],'y':tsne_f[:,1]})

print("Before:",X.shape)
print("After",df.shape)

display(df.head())
sns.scatterplot(x='x', y='y', hue=y['Label'],data=df)
plt.title("After Dimensionality Reduction")
plt.show()
```

Now, it gives us better understanding of the features

## 3. Build the model
First of all, we'll split the data into training and test sets

```python
# Split the data as 30% test and 70% training sets
X_train, X_test, y_train, y_test = train_test_split(breast_cancer.data, breast_cancer.target, test_size=0.3, random_state=34)
```

Now we can build our model and make predictions on test set

```python
# Initialize the Support vector classifier
svc = SVC(kernel="linear")

# Fit the SVC with training sets
svc.fit(X_train, y_train)

# Make predictions on test set
y_pred = svc.predict(X_test)
```

## 4. Evaluate the model's performance

Let's look at the accuracy score and classification report

```python
acc = accuracy_score(y_test, y_pred)
print("Accuracy:",acc)

print("\n Classification Report\n")
print(classification_report(y_test, y_pred))
```

It looks like our model works pretty well!
